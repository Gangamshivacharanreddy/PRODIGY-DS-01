```python
df=pd.read_csv('Demographics.csv')
df.head()
```

```python

# Calculate the total number of males and females
total_males = df['Males'].sum()
total_females = df['Females'].sum()
# Create a bar graph to compare the total number of males and females
data = {'Gender': ['Males', 'Females'], 'Count': [total_males, total_females]}
gender_df = pd.DataFrame(data)
# Plotting the bar graph
plt.figure(figsize=(8, 6))
sns.barplot(x='Gender', y='Count', data=gender_df, palette='viridis')
plt.title('Total Number of Males and Females')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Calculate the total population in urban and rural areas for both males and females
total_rural_population = df['Rural'].sum()
total_urban_population = df['Urban'].sum()
# Create a bar graph to compare the population distribution in urban and rural areas for both males and females
data = {
    'Area': ['Rural', 'Urban'],
    'Population': [total_rural_population, total_urban_population]
}
area_df = pd.DataFrame(data)
# Plotting the bar graph
plt.figure(figsize=(10, 6))
sns.barplot(x='Area', y='Population', data=area_df, palette='viridis')
plt.title('Population Distribution in Urban vs Rural Areas')
plt.xlabel('Area')
plt.ylabel('Population')
plt.show()
```

```python

# Calculate the number of females for every male in both rural and urban areas
rural_females_per_male = (df['Females'] * (df['Rural Population (%)'] / 100)) / (df['Males'] * (df['Rural Population (%)'] / 100))
urban_females_per_male = (df['Females'] * (df['Urban Population (%)'] / 100)) / (df['Males'] * (df['Urban Population (%)'] / 100))

# Sum the ratios
rural_females_per_male_avg = rural_females_per_male.mean()
urban_females_per_male_avg = urban_females_per_male.mean()

# Create a bar graph to represent the number of females for every male in both rural and urban areas
data = {
    'Area': ['Rural', 'Urban'],
    'Females per Male': [rural_females_per_male_avg, urban_females_per_male_avg]
}
area_gender_ratio_df = pd.DataFrame(data)

# Plotting the bar graph
plt.figure(figsize=(10, 6))
sns.barplot(x='Area', y='Females per Male', data=area_gender_ratio_df, palette='viridis')
plt.title('Number of Females for Every Male in Rural and Urban Areas')
plt.xlabel('Area')
plt.ylabel('Females per Male')
plt.ylim(0, 1.2)  # Set y-axis limit for better visualization
plt.show()
```
